```python
hr['terminated']=hr['Date of Termination'].fillna(0)
```

```python
table = pd.crosstab(hr['terminated'], hr['Pay Rate'])
chi2, p, dof, expected = chi2_contingency(table.values)
print('Chi-square Statistic %0.3f p_value %0.3f' % (chi2, p))
#No sig. ass'n b/w termination and pay rate
```

```python
table = pd.crosstab(hr['Zip'], hr['Pay Rate'])
chi2, p, dof, expected = chi2_contingency(table.values)
print('Chi-square Statistic %0.3f p_value %0.3f' % (chi2, p))
#No sig. ass'n b/w zip code and pay rate
```
